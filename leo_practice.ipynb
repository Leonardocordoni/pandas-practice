## Filtering

```python
bios.loc[bios["height_cm"] == 135]
```

```python
bios[bios["born_country"] == "FRA"].head(3)
```

```python
bios[bios['NOC'].isin(['France', 'Australia'])]
#negando
#bios[~bios['NOC'].isin(['France', 'Australia'])]
```

```python
bios.loc[
    (bios['born_country'] == 'FRA') &
    (bios['height_cm'].between(170, 190)) &
    (bios['died_date'].isnull())
]
```

```python
#com a query fica mais legivel
bios.query("born_country == 'FRA' and height_cm >= 170 and height_cm <= 190 and died_date.isnull()")
```

```python
bios.loc[lambda df: df['weight_kg'] > df['weight_kg'].mean()]
```

```python
bios.sort_values(by=["name", "born_date"])
```

```python
bios.sort_values(by=['born_country', 'height_cm'], ascending=[True, False])
```

```python
bios.nlargest(10, 'height_cm')
#bios.nsmallest(10, 'weight_kg')
```
